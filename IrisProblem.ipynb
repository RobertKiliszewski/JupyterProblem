```python
#All Necessary Imports for the plotting of data
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pat
```

```python
# https://stackoverflow.com/questions/3518778/how-to-read-csv-into-record-array-in-numpy

sepal_length = np.genfromtxt('Iris.csv', delimiter=',', usecols= (0))
sepal_width = np.genfromtxt('Iris.csv', delimiter=',', usecols= (1))
petal_length = np.genfromtxt('Iris.csv', delimiter=',', usecols= (2))
petal_width = np.genfromtxt('Iris.csv', delimiter=',', usecols= (3))
species = np.genfromtxt('Iris.csv', delimiter=',', usecols = (4), dtype = None)
```

```python
slope, variable = np.polyfit(sepal_length, sepal_width, 1)
print("Best fit is m = %f and c= %f" % (slope, variable))
```

```python
plt.plot(sepal_length, sepal_width, 'k.', label='Original data')
plt.plot(sepal_length, slope * sepal_length + variable, 'b-', label='Best fit: $%0.1f x + %0.1f$' % (slope,variable))
plt.xlabel('Sepal Length(Centimeters)')
plt.ylabel('Sepal Width(Centimeters)')
plt.title('Sepal Width Against Sepal Length Plot')
plt.rcParams['figure.figsize'] = (17.0, 7.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
different_species = np.genfromtxt('IRIS.csv', delimiter=',', dtype=None)
species_type = [x[4] for x in different_species]
```

```python
for i in range(0, len(different_species) -1 ):
    x, y = [different_species[i][0], different_species[i][1]]
    scale = 50.0
    # Give different colors for different flower types
    flower = species_type[i].decode("utf-8")
    color = " "
    if flower == "I.setosa":
        color = "red"
    elif flower == "I.versicolor":
        color = "green"
    elif flower == "I.virginica":
        color = "blue"

    pl.scatter(x, y, s=scale, c=color, alpha=1, edgecolor="white")
```

```python
#https://matplotlib.org/api/pyplot_api.html
#https://matplotlib.org/users/legend_guide.html

setosa = pat.Patch(color='red', label='Setosa')
versicolor = pat.Patch(color='green', label='Versicolor')
virginica = pat.Patch(color='blue', label='Virginica')
plt.legend(handles=[setosa, versicolor, virginica])

pl.title("The Iris Data Set")
pl.xlabel('sepal length')
pl.ylabel('sepal width')

pl.show()
```
